```python
class Clean:
  nltk.download('stopwords')


  def __init__(self, df, col):
    self.__df=df
    self.__col = col

  def __remove_url_html(self,col):
    import re
    import string
    col = col.lower()
    col = re.sub(r"http\S+", "", col)  # remove urls
    col = re.sub(r"<.*?>", "", col)    # remove html
    col = col.translate(str.maketrans("", "", string.punctuation))
    return col


  def apply_url_html(self):
    self.__df[self.__col]=self.__df[self.__col].apply(self.__remove_url_html)
    return self.__df


  def remove_stopwords_punctuation(self):
    import string
    stop_words = set(stopwords.words('english'))
    self.__df[self.__col]=self.__df[self.__col].apply(lambda x: ' '.join([word for word in x.split() \
                                                                  if (word not in (stop_words) and word not in list(string.punctuation))]))
    return self.__df

  def replace_short_hnads(self):
    for key, value in shorthand_dict.items():
      self.__df[self.__col]=self.__df[self.__col].str.replace(key, value)
    return self.__df


  def replace_emojee(self):
    self.__df[self.__col]=self.__df[self.__col].apply(lambda x: emoji.demojize(x))
    return self.__df

  def tokenize_lemiztize(self):
    nltk.download('wordnet')
    from nltk.stem import WordNetLemmatizer
    from nltk.tokenize import word_tokenize

    lemmatizer = WordNetLemmatizer()

    def process(self,col):
      tokens = word_tokenize(str(col))
      lemma=[lemmatizer.lemmatize(word) for word in tokens]
      return lemma
    self.__df[self.__col]=self.__df[self.__col].apply(lambda x: process(self,x))
    return self.__df

















```
